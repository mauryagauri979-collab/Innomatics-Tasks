```python
# Problem 1: Unique Words
sentence = "python is easy and python is powerful"
unique_words = set(sentence.split())
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Problem 2: Highest Salary
employees = {"Ravi":75000, "Anita":68000, "Kiran":72000}
highest = max(employees, key=employees.get)
print("Highest Salary:", highest, "-", employees[highest])
```

```python
# Problem 3: Max and Min
numbers = [45,22,89,10,66]
print("List:", numbers)
print("Max:", max(numbers))
print("Min:", min(numbers))
```

```python
# Problem 4: Count above 1000
prices = [450,1200,899,1500,300]
count = sum(1 for p in prices if p > 1000)
print("Products above 1000:", count)
```

```python
# Problem 5: Attendance Percentage
attendance = ['P','P','A','P','P']
percentage = attendance.count('P')/len(attendance)*100
print("Attendance Percentage:", percentage)
```

```python
# Problem 6: Remove Duplicate Phone Numbers
phone_numbers = [9876543210,9123456789,9876543210]
print("Unique phone numbers:", set(phone_numbers))
```

```python
# Problem 7: Character Frequency
text = "python"
freq = {}
for c in text:
    freq[c] = freq.get(c,0)+1
print(freq)
```

```python
# Problem 8: List to Tuple
lst = [10,20,30]
print("Tuple:", tuple(lst))
```

```python
# Problem 9: Check Key Exists
employees = {"Ravi":75000, "Anita":68000}
print("Employee exists" if "Ravi" in employees else "Employee does not exist")
```

```python
# Problem 10: Average Marks
marks = [70,75,80,65]
print("Average Marks:", sum(marks)/len(marks))
```
